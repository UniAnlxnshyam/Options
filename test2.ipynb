```python
import warnings
warnings.filterwarnings("ignore")
symbool = ['SP','EV']
callput = ['c','p']
expl = [1,3,6,12]

data_gen = DataExtractor(df,interpolator='linear')
for sym in symbool:
    for cp in callput:
        if cp=='c':
                mnl = [100,110,120]
        else:
            mnl =[100,90,80]
    
        for mn in mnl:
            for ex in expl:
                key = f'mn_{mn}_{cp}_{sym}_{ex}'

                if key in data_dict.keys():
                    continue
                print(f'generating {key}')
                df_op,msg_data = data_gen.gen_data(moneyness = mn, time_to_expiry = ex,callput = cp,bot =1,hold=ex,symbool=sym,start_date = 20100121)
                
                
                
                
                data_dict[key] =(df_op, msg_data)



```
